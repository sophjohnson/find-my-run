```python
#------------------------------------------ Add Weights to Adjacency List -----------------------------------------#

def findWeights(adjList, nodes):
    
    adjWeightedList = []
    # For each node
    for i, node in enumerate(adjList):
        nodeList = []  
        # For all neighbors of node, find distance and add to list
        for j in node:
            n = {}
            n['n'] = j
            n['w'] = math.sqrt((nodes[i]['y'] - nodes[j]['y']) ** 2 + (nodes[i]['x'] - nodes[j]['x']) ** 2)
            nodeList.append(n)
        adjWeightedList.append(nodeList)
        
    return adjWeightedList
```

```python
#------------------------------------------ Find Nodes and Adjacency List -----------------------------------------#

def chooseGraph(graphNum):   
    if graphNum == "graphOne":
        nodes, adjList = graphOne()
    elif graphNum == "graphTwo":
        nodes, adjList = graphTwo()
    elif graphNum == "graphThree":
        nodes, adjList = graphThree()
    elif graphNum == "graphFour":
        nodes, adjList = graphFour()
    else:
        nodes, adjList = graphFive()
    adjList = findWeights(adjList, nodes)
    return nodes, adjList
```

```python
#---------------------------------------------- Recursively Find Path ---------------------------------------------#

def findPathRec(start, v, dist, distDiff, adjList, paths, results):  
    
    # Check if path returns to start with correct distance
    if v == start and dist <= 0:
        if dist + distDiff >= 0:
            paths.append(results['path'])
        results['fnd'] = True
        return results    
    
    # If point has already been visited, return
    if v in results['marked']:
        return results
    
    # If you've traveled too far, return
    if dist <= 0:
        return results
    results['marked'].append(v)    

    # Add each neighbor to path 
    for neighbor in adjList[v]:
        
        # If path has been found, break
        if results['fnd'] and v != start:
            break
            
        # If start, set boolean to false
        elif v == start:
            results['fnd'] = False
            
        # Otherwise, add all neighbors
        results['path'].append(neighbor['n'])          
        results = findPathRec(start, neighbor['n'], dist - neighbor['w'], distDiff, adjList, paths, results) 
        results['path'] = results['path'][:-1]
    results['marked'] = results['marked'][:-1]  
    
    return results
```

```python
#------------------------------------------ Visualization Functions -----------------------------------------#

import time
import requests
import matplotlib.pyplot as plt

# Function to graph using adjacency list
def plot_graph_list(nodes,adjList):
    start_graph = time.time()
    for k in range(len(adjList)):
        for i in range(len(adjList[k])):
            ind1 = k
            for key in adjList[k][i].keys():
                if key == 'n':
                    ind2= adjList[k][i][key]
            plt.plot([nodes[ind1]['x'],nodes[ind2]['x']],[nodes[ind1]['y'],nodes[ind2]['y']],'b')
    plt.title('Total Graph')
    plt.show()
    end_graph = time.time()
    
# Function to plot routes using adjacency list
def plot_routes_list(nodes, adjList, path):
    start_routes = time.time()
    print("The black node represents the start/end point.")
    print("The arrows denote directionality of running path.")
    plt.rcParams['figure.figsize'] = [8, 8]
    for p in range(len(path)):
        for k in range(len(adjList)):
            for i in range(len(adjList[k])):
                ind1 = k
                for key in adjList[k][i].keys():
                    if key == 'n':
                        ind2= adjList[k][i][key]
                plt.plot([nodes[ind1]['x'],nodes[ind2]['x']],[nodes[ind1]['y'],nodes[ind2]['y']],'b')
        x = []
        y = []
        for entry in path[p]:
            x.append(entry['x'])
            y.append(entry['y'])
            plt.plot(x,y,'ro--')
        for i in range(len(x)-1):
            plt.arrow(x[i],y[i],(x[i+1]-x[i])/2,(y[i+1]-y[i])/2,head_width=0.1, head_length=0.2,color='r')
        plt.title("Route "+str(p+1))
        plt.plot(x[0],y[0],'ok')
        plt.show()
    end_routes = time.time()

# Function to plot main graph only
def plotMainGraph(graphNum):
    nodes, adjList = chooseGraph(graphNum)
    plot_graph_list(nodes, adjList)
```

```python
#---------------------------------------------- Main Function ---------------------------------------------#

def findRun(graphNum, x, y, dist, maxDist):
    
    # Find values based on user input
    nodes, adjList = chooseGraph(graphNum)
    distDiff = maxDist - dist

    # Function to find distances between node and starting point
    def calcDist(node, x, y):
        return math.sqrt((node['y'] - y) ** 2 + (node['x'] - x) ** 2)

    # Find closest starting node
    minIndex = 0
    minDist = calcDist(nodes[0], x, y)
    for index, node in enumerate(nodes):
        newDist = calcDist(node, x, y)
        if newDist <= minDist:
            minIndex = index
            minDist = newDist
    print("Start/End Point: ({}, {})".format(nodes[minIndex]['x'], nodes[minIndex]['y']))
    if minDist != 0:
        print("Note, the start point on this map is {} miles from the original position ({}, {}).".format(minDist, x, y))
    start = minIndex

    # Initialize variables for algorithm
    paths = []
    results = {
        'path' : [start],
        'marked' : [],
        'fnd' : False
    }

    # Call recursive function
    results = findPathRec(start, start, dist, distDiff, adjList, paths, results)

    # Set up path input for visualization
    visualization = []
    for i, path in enumerate(paths):
        path = [nodes[i] for i in path]
        visualization.append(path) 

    # Display results
    if not paths:
        print("No existing paths between {} and {} miles.".format(dist, maxDist))
    else:
        plot_graph_list(nodes, adjList)
        plot_routes_list(nodes, adjList, visualization)
        
    return paths
```

### Usage Instructions:
Use the dropdown menu to look at the various available maps. To have the app give you possible running paths of a certain distance, you must tell the app what map you would like to use, the x and y coordinates of your start location, your desired running distance, and the maximum distance you are willing to run. Enter this information in the text box at the bottom in the following manner: graph name, x coordinate, y coordinate, desired distance, maximum distance. For example, say I want to start on graphThree near point (1,2) and I want to run 6.5 miles, but I would be willing to run up to 7 miles. My input to the text box should be as follows:

        graphThree, 1, 2, 6.5, 7
        
Then hit enter. The app will return possible runs in the distance range specified. Simply repeat the process to try runs of different lengths on different maps.
