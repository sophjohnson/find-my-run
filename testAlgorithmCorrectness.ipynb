### Functions

```python
#------------------------------------------ Add Weights to Adjacency List -----------------------------------------#

def findWeights(adjList, nodes):
    
    adjWeightedList = []
    # For each node
    for i, node in enumerate(adjList):
        nodeList = []  
        # For all neighbors of node, find distance and add to list
        for j in node:
            n = {}
            n['n'] = j
            n['w'] = math.sqrt((nodes[i]['y'] - nodes[j]['y']) ** 2 + (nodes[i]['x'] - nodes[j]['x']) ** 2)
            nodeList.append(n)
        adjWeightedList.append(nodeList)
        
    return adjWeightedList
```

```python
#------------------------------------------ Find Nodes and Adjacency List -----------------------------------------#

def chooseGraph(graphNum):   
    if graphNum == 1:
        nodes, adjList = graphOne()
    elif graphNum == 2:
        nodes, adjList = graphTwo()
    elif graphNum == 3:
        nodes, adjList = graphThree()
    elif graphNum == 4:
        nodes, adjList = graphFour()
    else:
        nodes, adjList = graphFive()
    adjList = findWeights(adjList, nodes)
    return nodes, adjList
```

```python
#---------------------------------------------- Recursively Find Path ---------------------------------------------#

def findPathRec(start, v, dist, distDiff, adjList, paths, results):  
    
    # Check if path returns to start with correct distance
    if v == start and dist <= 0:
        if dist + distDiff >= 0:
            paths.append(results['path'])
        results['fnd'] = True
        return results    
    
    # If point has already been visited, return
    if v in results['marked']:
        return results
    
    # If you've traveled too far, return
    if dist <= 0:
        return results
    results['marked'].append(v)    

    # Add each neighbor to path 
    for neighbor in adjList[v]:
        
        # If path has been found, break
        if results['fnd'] and v != start:
            break
            
        # If start, set boolean to false
        elif v == start:
            results['fnd'] = False
            
        # Otherwise, add all neighbors
        results['path'].append(neighbor['n'])          
        results = findPathRec(start, neighbor['n'], dist - neighbor['w'], distDiff, adjList, paths, results) 
        results['path'] = results['path'][:-1]
    results['marked'] = results['marked'][:-1]  
    
    return results
```

### Output Verification

```python
#------------------------------------------ Find Run Function with No Display -------------------------------------------#

def findRunCorrect(graphNum, x, y, dist, maxDist):
    
    # Find values based on user input
    nodes, adjList = chooseGraph(graphNum)
    distDiff = maxDist - dist

    # Function to find distances between node and starting point
    def calcDist(node, x, y):
        return math.sqrt((node['y'] - y) ** 2 + (node['x'] - x) ** 2)

    # Find closest starting node
    minIndex = 0
    minDist = calcDist(nodes[0], x, y)
    for index, node in enumerate(nodes):
        newDist = calcDist(node, x, y)
        if newDist <= minDist:
            minIndex = index
            minDist = newDist
    #if minDist != 0:
        
    start = minIndex

    # Initialize variables for algorithm
    paths = []
    results = {
        'path' : [start],
        'marked' : [],
        'fnd' : False
    }

    # Call recursive function
    results = findPathRec(start, start, dist, distDiff, adjList, paths, results)

    # Set up path input for visualization
    visualization = []
    for i, path in enumerate(paths):
        path = [nodes[i] for i in path]
        visualization.append(path) 
        
    return paths, adjList
```

```python
#---------------------------------------------- Test Function ---------------------------------------------#
def testCorrectness(graphNum, x, y, dist, maxDist):
    
    paths, adjList = findRunCorrect(graphNum, x, y, dist, maxDist)
    counter = 1
    
    # check actual distance of each entry in paths
    for entry in paths:
        actualDist = 0;
        for i in range(len(entry)-1):
            for nodeEntry in adjList[entry[i]]:
                if nodeEntry['n'] == entry[i+1]:
                    actualDist = actualDist + nodeEntry['w']
        # if actual distance is outside of specified range, test fails
        if (actualDist < dist or actualDist > maxDist):
            print("Failure...")
            print("Route "+str(counter)+" Failed")
            print("Requested Distance: " + str(dist))
            print("Maximum Distance: " + str(maxDist))
            print("Actual Distance of Route: "+str(actualDist))
            return
        # track route number
        counter = counter + 1
    
    print("Success...")   
        
```
